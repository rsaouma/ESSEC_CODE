### Question 2

In this question you are asked to do 5 tasks:
 - load the file spx.csv which is the historical price of the S&P 500.
 - Plot the S&P time series - With the right Axis(price vs date), legend and title,
 - Plot the 50-days moving average in a different color - Annotate the line
 - Plot a band around the 50 days moving average of +/-100 point (as per the picture below)
 - Bonus: Color the area inside band with your choice of colour

```python
data = pd.read_csv("spx.csv")
band_size = 100
data.columns = ["date", "price"]
data.index = pd.to_datetime(data["date"])
data['50D MA'] = data.iloc[:,1].rolling(window=50).mean()

fig2, ax2 = plt.subplots(figsize=(15,10))
ax2.plot(data.index, data["price"], label="price")
ax2.plot(data.index, data["50D MA"], label = "50D MA", color = "red")
ax2.plot(data.index, data["50D MA"] + band_size, color="grey")
ax2.plot(data.index, data["50D MA"] - band_size, color="grey")
ax2.fill_between(data.index, data["50D MA"] - band_size,data["50D MA"] + band_size, alpha=0.5, color="grey")
ax2.set_xlabel("Year")
ax2.set_ylabel("Price")
ax2.legend(["SPX", "50D MA"])
```

### Question 3

Write a Python program to create bar plots with error bars on the same figure. Go to the editor
Sample Date
- Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824
- Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645
The code gives the output shown in the following screenshot:

![t_1](ex_3.png)

```python
sampleSize = 4
scores = (54.74, 42.35, 67.37, 58.24, 30.25)
errorBars = (4, 3, 4, 1, 5)
velocity = np.arange(0.18,sampleSize + 1)     

plt.bar(velocity, scores, 0.35, yerr=errorBars, color='red', error_kw=dict(ecolor='blue', lw=1.5, capsize=4, capthick=1.5))
plt.ylabel('Scores')
plt.xlabel('Velocity')
plt.title('Scores by Velocity')

plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.xlim(0,5)
plt.ylim(0,80)
plt.show()
```
